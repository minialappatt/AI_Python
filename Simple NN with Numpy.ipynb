```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#  y = w1x1 + w2x2
# x1 and x2 are features
# Y is the target
# W1 and W2 are the weights of the perceptron

# y is spend of a customer
# x1 is product 1
# x2 is product 2
# How many units of prod 1 ans 2 customer must purchase to achieve a sale of Y amount
```

```python
x1 = 4
x2 = 5
# x3 = 3.5
y = 10

# 10 = 4.w1+2.w2
```

```python
# learning rate
lr = 0.2
w1 = 0
w2 = 0
# w3 = 0
for epoch in range(500):
    y_pred = w1*x1 + w2*x2
    error = (y - y_pred)**2
    # dE/dW1 = 2(y-y_pred).(-x1)
    dEw1 = 2*(y - y_pred)*(-x1)
    dEw2 = 2*(y - y_pred)*(-x2)
    # dEw3 = 2*(y - y_pred)*(-x3)
    
    print("Epoch : ",epoch,"Value of w1 : ", w1, "Value of w2 : ", w2, "Error is : ", error, "Predicted is : ", y_pred)
    
    w1 = w1 - lr*dEw1
    w2 = w2 - lr*dEw2
    # w3 = w3 - lr*dEw3
    
```
